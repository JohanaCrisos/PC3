```python
import mapclassify
import numpy as np
K=5
#iniciamos con la columna %_of_GDP
#Se eligi√≥ el m√©todo FisherJenks
#Encuentra el esquema de clasificaci√≥n √≥ptimo minimizando la varianza dentro de las clases y maximizando la varianza entre clases. Su objetivo es crear las clases m√°s homog√©neas posibles.
intervaloGDP=mapclassify.FisherJenks(tablaunida['%_of_GDP'], k=K)
intervaloGDP
```

```python
#se repite la acci√≥n para cada columna
intervaloKW=mapclassify.FisherJenks(tablaunida['KW'], k=K)
intervaloBTU=mapclassify.FisherJenks(tablaunida['Btu/Person'], k=K)
intervaloVALUE=mapclassify.FisherJenks(tablaunida['Value'], k=K)
```

```python
#ahora agregaremos los intervalos en forma de columnas
tablaunida['intervalo_GDP']=intervaloGDP.yb
tablaunida['intervalo_KW']=intervaloKW.yb
tablaunida['intervalo_BTU']=intervaloBTU.yb
tablaunida['intervalo_VALUE']=intervaloVALUE.yb
tablaunida
```

```python
#Ahora podemos reorganizar las columnas para tener un mejor orden
nuevoorden=['Country','Region','%_of_GDP','intervalo_GDP','KW','intervalo_KW','Btu/Person','intervalo_BTU','Value','intervalo_VALUE']
tablaunida=tablaunida.reindex(columns=nuevoorden)
tablaunida
```
